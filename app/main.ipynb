## this code to make the csv file simple to use the inference in simple interface

```python
import pandas as pd

# Path to your input Excel file
input_file = 'final_datasetneed_column_only.xlsx'

# Load the Excel file into a DataFrame
df = pd.read_excel(input_file)

# Filter rows where the length of Concatinated_Transcript <= 160
df_filtered = df[df['Concatinated_Transcript'].str.len() <= 160]

# Display the filtered DataFrame
print(df_filtered)

# Path to save the filtered Excel file
output_file = 'filtered_transcripts.xlsx'

# Save the filtered DataFrame to a new Excel file
df_filtered.to_excel(output_file, index=False)
```
