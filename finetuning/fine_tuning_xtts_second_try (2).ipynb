## Change audio sample rate

```python
import time

def convert_sample_rate(audios_path, output_path, target_sample_rate):
  os.makedirs(output_path, exist_ok=True)
  audios = os.listdir(audios_path)
  start = time.time()
  for i, audio in enumerate(audios):
    if i % 100 == 0:
      current = time.time()
      print(f'Time take to process {i} audios: {current-start:.1f} seconds')
    audio_path = os.path.join(audios_path, audio)
    audio_output_path = os.path.join(output_path, audio)
    try:
      ffmpeg.input(audio_path).output(audio_output_path, ar=target_sample_rate).run(capture_stdout=True, capture_stderr=True) # .overwrite_output()
      print(f'Converted audio {i}')
      print(f"Successfully converted {audio_path} to {target_sample_rate} Hz")
    except ffmpeg.Error as e:
      print(f"Conversion failed. Error: {e.stderr.decode()}")


audios_path = '/content/gdrive/MyDrive/dataset'
output_path = '/content/gdrive/MyDrive/dataset22050'
target_sample_rate = 22050
convert_sample_rate(audios_path, output_path, target_sample_rate)
```

### Batch Size ###

* BATCH_SIZE is the amount of items being loaded into VRAM/memory at once.

* GRAD_ACCUM_STEPS is the amount of times we perform a forward pass with BATCH_SIZE amount of items before updating the parameters according to the SGD algorithm.

So a BATCH_SIZE of 2 and GRAD_ACCUM_STEPS of 32 would give an 'effective batch size' of 64 (i.e., 64 items considered per optimisation step).

The creators of XTTS-v2 recommend an effective batch size of 252 for proper training. (I found that reducing this to 126 gave me similar performance and faster fine-tuning, but we should probably listen to them.)

Your BATCH_SIZE will depend on how big your dataset items are and how much VRAM you have available. Make it as large as possible while ensuring that everything fits in VRAM and then make BATCH_SIZE*GRAD_ACCUM_STEPS==252.

### Train! ###

**Note on warnings:**

The trainer will print out warnings if it encounters items in your dataset where the text exceeds 166 chars or the length of your audio exceeds max_wav_length (It will also have problems if you have data items that are < ~0.2s long, which you won't want anyway). You should remove these from your dataset or re-think how you're creating your dataset.




```python
def clean_and_rename(samples):
    for sample in samples:
        original_path = sample['audio_file']

        # Check if the path ends with '.wav.wav'
        if original_path.endswith('.wav.wav'):
            # New path: remove one '.wav' and remove 'wavs/' from the path
            new_path = original_path.replace('/wavs/', '/').replace('.wav.wav', '.wav')

            # Rename the actual file
            if os.path.exists(original_path):
                os.rename(original_path, new_path)
                print(f"Renamed: {original_path} -> {new_path}")
            else:
                print(f"File not found: {original_path}")

            # Update the dictionary
            sample['audio_file'] = new_path
            sample['audio_unique_name'] = '#' + new_path.replace(sample['root_path'], '').lstrip('/')
    return samples

# Process both lists
train_samples = clean_and_rename(train_samples)
eval_samples = clean_and_rename(eval_samples)
train_samples
```

```python
trainer = Trainer(
    TrainerArgs(
        restore_path=None,
        skip_train_epoch=False,
        start_with_eval=START_WITH_EVAL,
        grad_accum_steps=GRAD_ACUMM_STEPS,
    ),
    config,
    output_path=OUT_PATH,
    model=model,
    train_samples=train_samples,
    eval_samples=eval_samples,
)
```

```python
trainer.fit()
```
