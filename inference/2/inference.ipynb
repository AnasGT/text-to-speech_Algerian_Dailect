```python
import os
import torch
import torchaudio
from TTS.tts.configs.xtts_config import XttsConfig
from TTS.tts.models.xtts import Xtts

CONFIG_FILE_PATH = '/content/gdrive/MyDrive/model4/config.json'
VOCAB_FILE_PATH = '/content/gdrive/MyDrive/model4/vocab.json'
MODEL_PATH = '/content/gdrive/MyDrive/model4'
SPEAKER_AUDIO_PATH = '/content/AymanCha7el_audio001_chunk13.wav'

print("Loading model...")
config = XttsConfig()
config.load_json(CONFIG_FILE_PATH)
model = Xtts.init_from_config(config)
model.load_checkpoint(config, checkpoint_dir=MODEL_PATH, use_deepspeed=True, vocab_path=VOCAB_FILE_PATH)
model.cuda()

print("Computing speaker latents...")
gpt_cond_latent, speaker_embedding = model.get_conditioning_latents(audio_path=[SPEAKER_AUDIO_PATH])
```

```python
import gradio as gr
import soundfile as sf
import os

def save_audio(audio_data):
    if audio_data is None:
        return "No audio recorded. Please record audio before submitting."

    try:
        audio_file, sample_rate = audio_data
        output_file = "recorded_audio.wav"
        sf.write('/content/'+output_file, audio_file, sample_rate)
        return f"Audio saved successfully as {output_file}"
    except Exception as e:
        return f"An error occurred: {str(e)}"

interface = gr.Interface(
    fn=save_audio,
    inputs=gr.Audio(sources=['microphone'], type="numpy"),
    outputs="text",
    title="Audio Recorder",
    description="Record an audio clip and save it to your device."
)

interface.launch()
```

```python
#FINAL WORKING VERSION WITH ALL FEATURES
import gradio as gr
import torchaudio
import torch

# Default audio path
default_audio_path = "/content/gdrive/MyDrive/finetuning_dataset/dataset/wavs/akhdar_audio_163_chunk_186.wav"

def inference(text, temperature, audio_path=None):
    print("Inference...")

    # Use default audio path if no audio is uploaded
    if audio_path is None:
        print(f"No audio uploaded. Using default audio path: {default_audio_path}")
        audio_path = default_audio_path

    # Generate conditioning latents
    print("Generating conditioning latents...")
    gpt_cond_latent, speaker_embedding = model.get_conditioning_latents(audio_path=[audio_path])

    out = model.inference(
        text,
        "ar",
        gpt_cond_latent,
        speaker_embedding,
        temperature=temperature,
    )
    # Save the audio to a temporary file
    temp_audio_file = "temp_audio.wav"
    torchaudio.save(temp_audio_file, torch.tensor(out["wav"]).unsqueeze(0), 24000)
    return temp_audio_file, temp_audio_file  # Return the file for both playback and download

iface = gr.Interface(
    fn=inference,
    inputs=[
        gr.Textbox(label="Enter text"),
        gr.Slider(0.0, 1.0, value=0.7, step=0.1, label="Temperature"),
        gr.Audio(type="filepath", label="Upload or Record Audio")  # This is where the user uploads or records audio
    ],
    outputs=[
        gr.Audio(label="Generated Audio"),  # Directly output audio
        gr.File(label="Download Audio")    # Provide file download option
    ],
    title="Egyptian Arabic Text-to-Speech",
    description="Enter Arabic text to generate speech. Upload or record audio to personalize.",
)

iface.launch()
```
